What's "Risk" in Bayesian A/B Testing?

## Risk is calculated as Expected Loss of making the wrong decision:

Expected Loss = Pr(Harm) x E[Magnitude of Harm | Harm]
Pr(Harm) is the probability that the treatment is worse than control, given the posterior 
E[Magnitude of Harm | Harm] is the avg amount by which the treatment is worse, only amony simulations where it is worse. 

## How to calculate risk with Beta Distribution

Suppose we model conversion rates as:
    Theta A ~ Beta (alpha-a, beta-a) (Control)
    Theta B ~ Beta (alpha-b, beta-b) (Treatment)
Then:
    1. Simulate posterior samples from both Beta distributions. 
    2. For each sample, compute delta = Theta-B - Theta-A
    3. Estimate risk as:
