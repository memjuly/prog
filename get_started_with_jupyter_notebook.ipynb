# Installation of Jupyter Notebook

To install Jupyter Notebook, you can use pip or Anaconda.

Using pip:

```bash
pip install jupyter
```

Using Anaconda:

Anaconda comes with Jupyter pre-installed. If not, use:

```bash
conda install jupyter
```

# Launching Jupyter Notebook

To launch Jupyter Notebook, open a terminal and run:

```bash
jupyter notebook
```

This will start the Jupyter server and open your browser to the notebook interface.

# Creating a New Notebook

In the Jupyter interface, click on "New" and select "Python 3" (or your kernel) to create a new notebook.

# Understanding the Notebook Interface

The Jupyter Notebook interface consists of:

- Menu bar: File, Edit, View, etc.
- Toolbar: Run, stop, save, etc.
- Cells: Where you write code or markdown.

# Running Code Cells

To run a code cell, select it and press Shift+Enter or click the Run button.

The output will appear below the cell.

# Markdown Cells for Documentation

To create a markdown cell, change the cell type to Markdown.

Use markdown syntax for formatting: # for headers, **bold**, etc.

# Saving and Exporting Notebooks

To save, press Ctrl+S or use File > Save.

To export, use File > Download as > HTML, PDF, etc.
